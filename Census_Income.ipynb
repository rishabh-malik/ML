```python
import numpy as np
from sklearn import preprocessing, cross_validation, neighbors,svm
import pandas as pd
```

```python
df=pd.read_table('census_income.data',sep=',')
```

```python
df.columns=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','y']
df.head()
```

```python
df.convert_objects(convert_numeric=True)
df.fillna(0,inplace=True)
df.head()
```

```python
def handle_non_numeric_data(df):
    columns=df.columns.values
    
    for column in columns:
        text_digit_vals={}
        def convert_to_int(val):
            return text_digit_vals[val]
        
        if df[column].dtype != np.int64 and df[column].dtype != np.float64:
            column_contents = df[column].values.tolist()
            unique_elements = set(column_contents)
            x = 0
            for unique in unique_elements:
                if unique not in text_digit_vals:
                    text_digit_vals[unique] = x
                    x+=1

            df[column] = list(map(convert_to_int, df[column]))

    return df

df=handle_non_numeric_data(df)
df.head()
```

```python
x=np.array(df.drop('y',1).astype(float))
x=preprocessing.scale(x)
y=np.array(df['y'])
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, Y_train, Y_test = tts(x,y, test_size=0.2)
```

```python
clf=svm.SVC()
clf.fit(X_train,Y_train)
```

```python
accuracy=clf.score(X_test,Y_test)
accuracy
```
